```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('breach_report.csv')
df.head()
```

```python
df.describe()
```

```python
base_color = sns.color_palette()[1]
year_order = df['Year'].value_counts().index
state_order = df['State'].value_counts().index
attack_order = df['Type of Breach'].value_counts().index
```

## Attacks per year

```python
plt.figure(figsize=(20,8))
sns.countplot(data=df,y='Year',color=base_color,order=year_order)
plt.xticks(rotation=45);
```

## Attacks per State

```python
plt.figure(figsize=(20,8))
sns.countplot(data=df,x='State',color=base_color,order=state_order)
plt.xticks(rotation=45);
```

```python
plt.figure(figsize=(20,8))
sns.countplot(data=df,x='State',color=base_color,order=state_order)
plt.xticks(rotation=45);
```

## Attack density per year

```python
plt.figure(figsize=(20,8))
plt.scatter(data = df, y = 'Individuals Affected', x = 'Year', alpha=1/2)
plt.xticks(rotation=15);
```

```python
plt.hist2d(data = df, y = 'Individuals Affected', x = 'Year',cmin=0.5,cmap = 'viridis_r')
plt.colorbar();
```

### Positive Relationship between attacks and years

```python
sns.violinplot(data=df, y='Type of Breach', x='Year', color=base_color, innner=None,
              order=attack_order)
plt.xticks(rotation=15);
```

```python
plt.figure(figsize=(20,8))
sns.countplot(data = df, x = 'Year', hue = 'Type of Breach',color=base_color,order=year_order);
```

```python
plt.figure(figsize=(20,8))
ax = sns.barplot(data = df, x = 'Year', y = 'Individuals Affected', 
                hue = 'Type of Breach',palette='viridis',order=year_order)
# ax.legend(loc = 8, ncol = 3, framealpha = 1, title = 'Type_of_Breach')
plt.xticks(rotation=15);
```

```python
plt.hist2d(data = df, y = 'Individuals Affected',x = 'Year',cmin=0.1,cmap = 'vlag')
plt.colorbar(label='Frequency')
plt.title('Individuals Affected by cyber-attacks per year')
plt.xlabel('Year')
plt.ylabel('Individuals Affected')
plt.yscale('log');
```
